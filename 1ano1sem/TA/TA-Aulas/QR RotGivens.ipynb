```python
#FUNCIONA PARA TUDO
A = matrix([[0.8147, 0.0975, 0.1576],
            [0.9058, 0.2785, 0.9706],
            [0.1270, 0.5469, 0.9572],
            [0.9134, 0.9575, 0.4854],
            [0.6324, 0.9649, 0.8003],
            ])



def copy(to_copy):
    l = []
    for line in to_copy:
        l.append(list(line))
    return matrix(l)


def prod_matrizes(L):
    
    result = L[0]
    for i in range(1,len(L)):
        result *= L[i]
    return result
    
    
def rot_givens_fact_QR(A):
    n = A.nrows()
    m = A.ncols()
    R = copy(A)
    In = identity_matrix(RR, n)
    Plist = []
    
    
    for i in range(0,m):
        for j in range(i+1,n):
            cos = R[i,i] / sqrt( (R[i,i]^2) + (R[j,i]^2) )
            sin = R[j,i] / sqrt( (R[i,i]^2) + (R[j,i]^2) )
            P = copy(In)
            P[i,i] = P[j,j] = cos
            P[j,i] = -sin
            P[i,j] = sin
            R = P*R
            Plist.append(P.transpose())
    
    Q = prod_matrizes(Plist)
    
    return Q, R 


Q, R = rot_givens_fact_QR(A)


#Q ortogonal: qT = q-1 para verificar fazer   q*qT == In
#R triang. sup. s√≥ zeros abaixo da diagonal
#Q*R tem de ser igual a matriz original


print(f"Q:{Q} \n\n R:{R} \n\n QR: {Q*R} \n\n QxQT: {Q*Q.transpose()}")




            
            
            
            
        
    
    
```
