```python

#definir k! mod n
#resto da divisao de k! por n

def fact_mod(k, n):
    Zn = IntegerModRing(n)
    return Zn(factorial(k))

#acho q e equivalente a outra

print(fact_mod(35,89))
```

```python
#Teorema Numeros Primos
# funcao pi

#   pi : R+ -> N U {0}

#        n  -> numero de primos entre 1 e n inclusive

def pi(n):
    lt = [p for p in range(1, floor(n)+1) if is_prime(p)]
    print(lt)
    return len(lt)

pi(7.8)

#ja existe definida chama-se prime_pi
```
