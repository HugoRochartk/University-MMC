```python
import numpy as np
import numpy.linalg as linalg
from matplotlib import pyplot as plt
```

```python
mu_vec = np.array([0,0])
cov_mat = np.array([[1,0.8], [0.8,1]])
amostra = np.random.multivariate_normal(mu_vec, cov_mat, 500).T
x, y = amostra[0], amostra[1]

plt.plot(x, y, 'o')
plt.show()
```

```python
med_x = np.mean(x)
med_y = np.mean(y)
med = [med_x, med_y]
```
